Q1:
Create a program that asks user to enter their birthday and then prints out the following pieces of information in separate lines:
1. What day of the week they were born
2. How many days to go before their next birthday
3. What is their current age?

```
import pickle
import os
from datetime import date, time, datetime, timedelta

EMPLOYEE_LIST_FILE = 'employees.bin'
LOG_FILE_NAME = 'employees_log.txt'

def clock_in(employees):
    name = input('[CLOCKING IN] Enter Employee name: ')
    current_time = datetime.now()
    employee_found = False
    for employee in employees:
        if employee[0] == name:
            employee_found = True
            employee[1] = True
            employee[2] = current_time
            with open(LOG_FILE_NAME, 'a+') as log_file:
                log_file.write(f'{name} clocked in at {current_time}')
            print('Data is updated')
    if not employee_found:
        print('Employee Name not found.')
    

def clock_out(employees):
    name = input('[CLOCKING OUT] Enter Employee name: ')
    current_time = datetime.now()
    employee_found = False
    for employee in employees:
        if employee[0] == name:
            employee_found = True
            employee[1] = False
            employee[2] = None
            with open(LOG_FILE_NAME, 'a+') as log_file:
                log_file.write(f'{name} clocked out at {current_time}')
            print('Data is updated')
    if not employee_found:
        print('Employee Name not found.')


def add_employee(employees):
    name = input('Enter New Employee name: ')
    employees.append([name, False, None])
    with open(EMPLOYEE_LIST_FILE, 'wb') as file:
        pickle.dump(employees, file)
    
def list_employees(employees):
    for employee in employees:
        print(employee[0])

def main():
    # Read employee list from file and load the data.
    if os.path.exists(EMPLOYEE_LIST_FILE):
        with open(EMPLOYEE_LIST_FILE, 'rb') as file:
            employee_list = pickle.load(file)
    else:
        employee_list = []
    
    while True:
        print('1. Clock In')
        print('2. Clock Out')
        print('3. Add Employee')
        print('4. List Employees')
        print('0. End Program')
        
        choice = input('Enter Choice: ')
        if choice == '1':
            clock_in(employee_list)
        elif choice == '2':
            clock_out(employee_list)
        elif choice == '3':
            add_employee(employee_list)
        elif choice == '4':
            list_employees(employee_list)
        elif choice == '0':
            break
        else:
            print('Invalid Choice.')
    

if __name__ == '__main__':
    main()
```
